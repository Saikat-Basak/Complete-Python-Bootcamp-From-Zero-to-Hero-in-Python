```python
work_hours = [('Abby',100),('Billy',400),('Cassie',800),('ssie',1000),('ie',400)]
```

```python
def employee_check(work_hours):
    current_max=0
    employee_name=''
    for name,hours in work_hours:
        if hours>current_max:
            current_max=hours
            employee_name=name
    
    return (employee_name,current_max)
```

```python
def player_guess():
    guess=''
    while guess not in ['0','1','2']:
        guess=input("pick a number: 0, 1 or 2")
    return int(guess)
```

```python
new_list=shuffle_list([' ', 'O', ' '])
num=player_guess()
if new_list[num]=='O':
    print('your guess is correct')
else:
    print('Try again..better luck next time')
    print(new_list)
```

```python
def myfunc(word):
    new_list=''
    for num,letter in enumerate(word):
        if num%2==0:
            new_list+=letter.lower()
        else:
            new_list+=letter.upper()
    return new_list
```

```python
myfunc('Hello world')
```

```python
def myfunc(word):
    new_word=[]
    for i in range(len(word)):
        if i%2==0:
            new_word.append(word[i].lower())
        else:
            new_word.append(word[i].upper())
    return ''.join(new_word)
```

```python
myfunc('Hello world')
```
